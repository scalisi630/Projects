```python
import numpy

stocks_list = []


def helper_0():
   a = input("Please enter the list of stock prices you would like analyzed     ")
   b = a.split(',')
   c = [int(x) for x in b]
   for element in c:
     stocks_list.append(element)
   main()


def helper_1(): 
  n = 1
  for i in stocks_list:
        print('Day',''' ''', n, ''': $''',i, sep='')
       
        n += 1
  main()


def helper_2():   
  x = numpy.average(stocks_list)
  c = round(x,2)
  print(f'The average price rounded to the nearest cent is ${c}.')
  
  
  main()

def helper_3():  
  x = numpy.std(stocks_list)
  c = round(x, 2)
  print(f'The standard deviation is {c} rounded to two decimal places')
  
  main()


def helper_4(): 
  min_v = min(stocks_list)
  min_i = stocks_list.index(min_v) + 1
  print(f'The minimum value is ${min_v} on day {min_i}.') 
  main()
  
def helper_5():
  
  max_v = max(stocks_list)
  max_i = stocks_list.index(max_v) + 1
  print(f'The maximum value is ${max_v} on day {max_i}.')
  main()


def helper_6():
  max_profit = []
  for b in range(len(stocks_list)):
    for s in range(b, len(stocks_list)):
      print('Buy:', b+1, 'Sell:', s+1, 'Profit:', stocks_list[s] - stocks_list[b])
      max_profit.append((stocks_list[s] - stocks_list[b], b+1, s+1))
    print()
  max_profit = max(max_profit)
  print(f"To obtain a maximum profit of ${max_profit[0]}, buy on day {max_profit[1]}, and sell on day {max_profit[2]}.")


  main()



def main():
  print('''
(0) Input a new list
(1) Print the current list
(2) Find the average price
(3) Find the standard deviation
(4) Find the minimum and its day
(5) Find the maximum and its day
(6) Your Spartan Investment Plan
(9) Quit
Enter your choice: ''')
   
    
  try:
    f = int(input())
   
    if f == 0:
     helper_0()
    elif f==1:
      helper_1()
    elif f ==2:
      helper_2()
    elif f==3:
       helper_3()
    elif f==4:
       helper_4()
    elif f ==5:
      helper_5()
    elif f ==6:
      helper_6()
    elif f ==9:
      print('See you soon')
      
      
    else:
      print(f'Please input a valid number, {f} is not a valid number')
      main()
  except ValueError:
    print('Please input a valid number.')
    main()
```
